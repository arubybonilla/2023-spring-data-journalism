**Note about social security increases:**

Each year the U.S. government increases the amount of social security retired people receive. That increase is based on the annual inflation rate  and the political climate (or party in control). I have indexed the numbers so 1995 is 100, and each year after that the number reflects a percentage increase since 1995. So between 1995 at 100 and 1996 at 102.6 reflects a 2.6 percent increase. And between 1995 and 1997 reflects a 5.6 percent increase.


```python
import pandas as pd
```

```python
df = pd.read_csv("cable-bill-only.csv")
df
```

```python
df.info()
```

```python
df.columns = ["year", "cable bill", "median income", "social security"]
df
```

```python
## fixed value of cable bill
old_cable_bill = df["cable bill"].iat[0]
old_cable_bill
```

```python
## percent change of cable bill
df["bill_pct_chg"] = ((df["cable bill"] - old_cable_bill) / old_cable_bill) * 100
df
```

```python
## fixed value of median income
old_median_income = df["median income"].iat[0]
old_median_income
```

```python
## percent change of median income
df["income_pct_chg"] = ((df["median income"] - old_median_income) / old_median_income) * 100
df
```

```python
## fixed value of social security
old_ss = df["social security"].iat[0]
old_ss
```

```python
## percent change of median income
df["ss_pct_chg"] = ((df["social security"] - old_ss) / old_ss) * 100
df
```

```python
## mean, median
df[["cable bill", "median income"]].agg(["mean", "median"])
```

```python
## cable bill red; median income blue, and social security gray.

df.plot (x = "year",
        y = ["bill_pct_chg", "income_pct_chg", "social security"],
        style = {"bill_pct_chg":"red",
                    "income_pct_chg":"blue",
                        "social security":"gray"})
```

```python
df.plot (x = "year",
        y = ["bill_pct_chg", "income_pct_chg", "ss_pct_chg"],
          style = {"bill_pct_chg":"red",
                    "income_pct_chg":"blue",
                        "ss_pct_chg":"gray"})
```

```python
df.plot (x = "year",
        y = ["income_pct_chg", "ss_pct_chg"],
          style = {"income_pct_chg":"blue",
                        "ss_pct_chg":"gray"})
```
